```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
svgd1 = np.loadtxt("results/svgd_covtype_2epochs_100particles.csv", delimiter=",")
```

```python
sgld1 = np.loadtxt("results/sgld_covtype_2epochs_100particles.csv", delimiter=",")
```

```python
fig = plt.figure(figsize=(10,6), dpi=100)
```

```python
fig = plt.figure(figsize=(6,6), dpi=100)
plt.plot(np.arange(1, svgd1.shape[0]+1), svgd1, marker="o", color='red', label="SVGD")
plt.plot(np.arange(1, svgd1.shape[0]+1), sgld1, marker="|", color='blue', label="SGLD")
plt.legend()
plt.xlabel("iterations/1000")
plt.ylabel("Test accuracy")
plt.xlim([1,18])
plt.xticks(range(1,19), np.arange(1,svgd1.shape[0]+1))
```
